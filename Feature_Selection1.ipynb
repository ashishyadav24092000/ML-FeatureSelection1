## Handling the missing values

```python
## lets handle Categorical features which are missing
feature_nan = [feature for feature in data_hp.columns if data_hp[feature].isnull().sum()>1 and data_hp[feature].dtypes=='O']

for feature in feature_nan:
    print("{}:{} %missing values".format(feature,np.round(data_hp[feature].isnull().mean(),4)))
```

```python
data_hp[feature_nan].isnull().sum()
```

So our categorical features have the null/nan values as shown above.

```python
## Replacing missing categorical Values with a new label
def replace_nan_feature(data_hp,feature_nan):
    data = data_hp.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    return data

##function calling
dataset = replace_nan_feature(data_hp,feature_nan)
dataset[feature_nan].isnull().sum()
```

So finally the nan values from the categorical features has been removed.

```python
dataset.head(10)
```

```python
## Numerical Feature that contains nan/Null values
numerical_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']
print(numerical_with_nan)
print("-"*45)
## displaying the nan value along with their missing percentage
for feature in numerical_with_nan:
    print("{}:{} %missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=dataset[feature].median()
    
    ## create a new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numerical_with_nan].isnull().sum()
```

So finally the nan values from the numerical features also has been removed.

```python
dataset.head(10)
```

```python
## Handling Temporal Variables (Date time variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature] = dataset['YrSold']-dataset[feature]  ## it'll give total years for the year features
```

```python
dataset.head()
```

```python
## So particulary our year features would now look something like this
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

## NOTE --> Since the Numerical distributions are skewed (We can see from previous scatter plots of continuous variables). So we will perform log Normal Distribution.

```python
dataset.head(n=10)
```

```python
num_feature = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice'] ## num features with non zero values

for feature in num_feature:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset.head(15)
```

## Now Handling Rare Categorical Features

i.e. We will remove categorical variables that are present less than 1% of the observations.

```python
## capture the categorical features
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes=='O']
categorical_feature
```

```python
len(dataset)
```

```python
## DataFrame.isin(values) :: 
    ## --> Whether each element in the DataFrame is contained in values.


for feature in categorical_feature:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.01].index ## It stores name of features as per the condition specified
    dataset[feature] = np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(50)
```

## Converting all categorical labels into numerical labels so that it does not bother us while Applying Machine Learning Algorithms::

```python
for feature in categorical_feature:
    labels_ordered = dataset.groupby(feature)['SalePrice'].mean().sort_values().index
    print(labels_ordered)
```

```python
for feature in categorical_feature:
    labels_ordered = dataset.groupby(feature)['SalePrice'].mean().sort_values().index
    labels_ordered = {k:i for i,k in enumerate(labels_ordered,start=0)}
    print(labels_ordered)
    ## dataset[feature] = dataset[feature].map(labels_ordered)
```

```python
for feature in categorical_feature:
    labels_ordered = dataset.groupby(feature)['SalePrice'].mean().sort_values().index
    labels_ordered = {k:i for i,k in enumerate(labels_ordered,start=0)}
    dataset[feature] = dataset[feature].map(labels_ordered)
```

```python
dataset.head(n=50)
```

```python
### Apply Feature Selection
# first, I specify the Lasso Regression model, and I
# select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then I use the selectFromModel object from sklearn, which
# will select the features which coefficients are non-zero


selected_features_model = SelectFromModel(Lasso(alpha=0.005,random_state = 0))  # remember to set the seed, the random state in this function
selected_features_model.fit(x_train,y_train)
```
